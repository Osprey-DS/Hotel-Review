```python
df = pd.read_csv('Data Analyst - Test Data - US.csv')
```

```python
df.head()
```

```python
df['Review'] = df['Review'].astype(str)
```

```python
import re
import nltk
import string
from nltk.corpus import stopwords
```

```python
def remove_punc(message):
    text_punc_removed = [char for char in message if char not in string.punctuation]
    text_punc_removed_join = ''.join(text_punc_removed)

    return text_punc_removed_join.lower()
```

```python
df['Review_out_punct'] = df['Review'].apply(remove_punc)
```

```python
def get_word_list(fname):
    with open(fname) as fh:
        return list(fh.read().split())
```

```python
positive = get_word_list('positive_word.txt')
positive
```

```python
df['Review split'] = df['Review_out_punct'].apply(lambda x: x.split())
```

```python
df.head()
```

```python
def sentiment(row):
    
    for word in row:
        if word in positive:
            return ("Positive")
        elif word in negative:
            return ('Negative')
        else:
            return ('Nutral')
```

```python
df['Sentiment'] = df['Review split'].apply(sentiment)
```
